```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()


```

```
from google.colab import drive
drive.mount('/content/drive')
```

```

sgd = pd.read_csv('/content/drive/My Drive/DL_Plot_CSV/ResNet/SGD_ResNet.csv')
adagrad = pd.read_csv('/content/drive/My Drive/DL_Plot_CSV/ResNet/AdaGrad_ResNet.csv')
adam = pd.read_csv('/content/drive/My Drive/DL_Plot_CSV/ResNet/Adam_ResNet.csv')
amsgrad = pd.read_csv('/content/drive/My Drive/DL_Plot_CSV/ResNet/AMSGrad_ResNet.csv')
adabound = pd.read_csv('/content/drive/My Drive/DL_Plot_CSV/ResNet/AdaBound_ResNet.csv')
amsbound = pd.read_csv('/content/drive/My Drive/DL_Plot_CSV/ResNet/AMSBound_ResNet.csv')
```

```
fig,axs = plt.subplots(2,2,figsize=(13,9))

fig.suptitle('Results for ResNet34',fontsize=16)
axs[0,0].plot(range(200),sgd['train_acc'],label='SGDM')
axs[0,0].plot(range(200),adagrad['train_acc'],label='AdaGrad')
axs[0,0].plot(range(200),adam['train_acc'],label='Adam')
axs[0,0].plot(range(200),amsgrad['train_acc'],label='AMSGrad')
axs[0,0].plot(range(200),adabound['train_acc'],linestyle='--',label='AdaBound')
axs[0,0].plot(range(200),amsbound['train_acc'],linestyle='--',label='AMSBound')

axs[0,0].set_title('train accuracy')
axs[0,0].legend(loc='lower right')

#############################################


axs[0,1].plot(range(200),sgd['val_acc'],label='SGDM')
axs[0,1].plot(range(200),adagrad['val_acc'],label='AdaGrad')
axs[0,1].plot(range(200),adam['val_acc'],label='Adam')
axs[0,1].plot(range(200),amsgrad['val_acc'],label='AMSGrad')
axs[0,1].plot(range(200),adabound['val_acc'],linestyle='--',label='AdaBound')
axs[0,1].plot(range(200),amsbound['val_acc'],linestyle='--',label='AMSBbound')
axs[0,1].set_title('test accuracy')
axs[0,1].legend(loc='lower right')


#########################################################
axs[1,0].plot(range(200),sgd['train_loss'],label='SGD')
axs[1,0].plot(range(200),adagrad['train_loss'],label='AdaGrad')
axs[1,0].plot(range(200),adam['train_loss'],label='Adam')
axs[1,0].plot(range(200),amsgrad['train_loss'],label='AMSGrad')
axs[1,0].plot(range(200),adabound['train_loss'],linestyle='--',label='AdaBound')
axs[1,0].plot(range(200),amsbound['train_loss'],linestyle='--',label='AMSBound')


axs[1,0].set_title('train loss')
axs[1,0].legend(loc='upper right')



##############################################

axs[1,1].plot(range(200),sgd['val_loss'],label='SGDM')
axs[1,1].plot(range(200),adagrad['val_loss'],label='AdaGrad')
axs[1,1].plot(range(200),adam['val_loss'],label='Adam')
axs[1,1].plot(range(200),amsgrad['val_loss'],label='AMSGrad')
axs[1,1].plot(range(200),adabound['val_loss'],linestyle='--',label='AdaBound')
axs[1,1].plot(range(200),amsbound['val_loss'],linestyle='--',label='AMSBound')


axs[1,1].set_title('test loss')
axs[1,1].legend(loc='upper right')
plt.savefig('resNet34_result_plot.png')
plt.show()
```
